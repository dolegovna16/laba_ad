```python
import pandas as pd
import numpy as np
import time

# Таймер для зчитування pandas
start = time.time()
pd_df = pd.read_csv('input.txt', sep=';', na_values='?')
print("Час зчитування pandas:", time.time() - start, "секунд")
```

```python
start = time.time()
types = [("Date", "U10"), ("Time", "U8"), ("Global_active_power", "float64"),
         ("Global_reactive_power", "float64"), ("Voltage", "float64"),
         ("Global_intensity", "float64"), ("Sub_metering_1", "float64"),
         ("Sub_metering_2", "float64"), ("Sub_metering_3", "float64")]
np_df = np.genfromtxt("input.txt", missing_values=["?", np.nan],
                      delimiter=';', dtype=types, encoding="UTF=8", names=True)
print("Час зчитування numpy:", time.time() - start, "секунд")
```

Завдання 3 (сила струму 19-20, ті записи, де Sub_metering_2 > Sub_metering_3)

```python
start = time.time()
pd_current = pd_df[(pd_df['Global_intensity'] >= 19) & (pd_df['Global_intensity'] <= 20)]
pd_rows = pd_current['Sub_metering_2'] > pd_current['Sub_metering_3']
print(pd_current[pd_rows])
print("Час складної фільтрації pandas:", time.time() - start, "секунд")
```

```python

start = time.time()
np_current = np_df[(np_df['Global_intensity'] >= 19) & (np_df['Global_intensity'] <= 20)]
np_rows = np_current['Sub_metering_2'] > np_current['Sub_metering_3']
print(np_current[np_rows])
print("Час складної фільтрації numpy:", time.time() - start, "секунд")

```

Завдання 4 Обрати випадковим чином 500000 записів (без повторів елементів
вибірки), для них обчислити середні величини усіх 3-х груп
споживання електричної енергії

```python
start = time.time()
pd_sample = pd_df.sample(n=500000)
np_sample_indices = np.random.choice(len(np_df), size=500000, replace=False)
np_sample = np_df[np_sample_indices]
avg_1 = pd_sample[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].mean()
avg_2 = f"Sub_metering_1 : {np.nanmean(np_sample['Sub_metering_1'])}"
avg_3 = f"Sub_metering_2 : {np.nanmean(np_sample['Sub_metering_2'])}"
avg_4 = f"Sub_metering_3 : {np.nanmean(np_sample['Sub_metering_3'])}"
print(avg_1)
print(avg_2)
print(avg_3)
print(avg_4)
print("Час підрахунку середніх значень:", time.time() - start, "секунд")
```

Завдання 5 Обрати ті записи, які після 18-00 споживають понад 6 кВт за
хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другої половини.

```python
start = time.time()
groups = pd_df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']]
pd_after_18 = pd_df.loc[(pd_df['Time'] >= '18:00:00') &
                        (pd_df['Global_active_power'] > 6) &
                        (groups.max(axis=1) == pd_df['Sub_metering_2'])]
print("Час групування після 18:00:", time.time() - start, "секунд")
```

А тепер те саме тільки через np

```python
mask1 = np_df['Time'] >= '18:00:00'
mask2 = np_df['Global_active_power'] > 6

max_matr = np.stack([np_df['Sub_metering_1'], np_df['Sub_metering_2'], np_df['Sub_metering_3']])
max_ind = np.argmax(max_matr, axis=0)
mask3 = max_ind == 1

np_after_18 = np_df[mask1 & mask2 & mask3]
```
